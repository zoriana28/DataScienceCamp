```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    np.random.seed(345)
    a = np.random.randint(1,100,20)
    deep = np.argmin(a)
    rightsubarray = a[deep: len(a)+1]
    for i in range(deep, len(a)):
        if i < len(a) - 1 and a[i] == max(rightsubarray):
           r = i
           break
    leftsubarray=a[0:deep+1]
    for i in range(deep, 0, -1):
        if i > 0 and a[i]==max(leftsubarray):
            l = i
            break
    b = np.array(a[l:r + 1])
    x = np.arange(l, r + 1)
    plt.plot(a, color="blue")
    plt.plot(x, b, color="red")
    plt.show()

if __name__ == "__main__":
    main()
```
